1. Creating SQLite DB

```python
import pandas as pd
import sqlite3
df = pd.read_csv("../data/ecommerce_sales_cleaned.csv")

conn = sqlite3.connect("ecommerce_sales.db")

df.to_sql(
    name="sales",
    con=conn,
    if_exists="replace",
    index=False
)

conn.close()
```

```python
# Verifying data insertion

conn = sqlite3.connect("ecommerce_sales.db")

query = "SELECT COUNT(*) FROM sales;"
pd.read_sql(query, conn)
```
